##### Model Training and Summary

```python
import time
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import mean_squared_error
```

```python
def model_run(df:pd.DataFrame,target_columns:list,algos:list,n:int)->pd.DataFrame:
  """
  Train and evaluate models for regression task using specified algorithms.

  Parameters:
      df (pd.DataFrame): The DataFrame containing the dataset.
      target_columns (list): A list of target columns to predict.
      algos (list): A list of tuples, where each tuple contains an algorithm name and an instance of the algorithm.
      n (int): The number of rows to consider from the dataset.
                This can be used to limit and check the time and resource before running large dataset.

  Returns:
      result: A DataFrame containing the feature name, model name, and Mean Squared Error (MSE) for each combination of feature and algorithm.
  """

  # Identifying numeric and categorical features
  numeric_features=df.select_dtypes(include=['float64','int64']).columns
  categorical_features=df.select_dtypes(include=['object']).columns

  # Initializing transformers
  numeric_transformer=StandardScaler()
  categorical_transformer=OneHotEncoder(handle_unknown='ignore')

  # List to store results
  loop_results:list=[]

  print(time.ctime())

  # Iterating over features
  for feature in df.drop(columns=target_columns).columns:

    temp_df=df[[feature]+target_columns]

    # Filtering the dataset based on specified number of rows
    if n>0:
      temp_df=temp_df.head(n)

    # Filtering the dataset based on placeholder value (useful as this takes the true distribution before running the models)
    temp_df=temp_df[temp_df[feature]!=-999999]
    X=temp_df[[feature]]
    Y=temp_df[target_columns]

    # If we have a single Target to predict then models will expect numpy series and not a dataframe
    if len(target_columns):
      Y=Y.squeeze()

    # Preprocessing features
    if feature in numeric_features:
      X=numeric_transformer.fit_transform(X)
    else:
      X=categorical_transformer.fit_transform(X)

    # Splitting data into train and test sets
    X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=123)
    del X,Y

    # Converting to dense array if required
    if not isinstance(X_train,np.ndarray):
      X_train=X_train.toarray()
      X_test=X_test.toarray()

    # Training models and storing results (main loop)
    for algo_name,algo in algos:
      algo.fit(X_train,y_train)
      loop_results.append((
          feature,
          algo_name,
          mean_squared_error(y_test,algo.predict(X_test)),
          ))
      print(algo_name,feature,time.ctime())
      print('='*50)
      del algo_name,algo
  # Model Results
  result_df=pd.DataFrame(loop_results,columns=['Feature','Model','MSE']).sort_values(by=['MSE'],ascending=[True])
  del loop_results
  return result_df
```

```python
def display_top_results(df,n,algos)->None:
  """
  Display the top results (features with lowest MSE) for each algorithm.

  Parameters:
      df (pd.DataFrame): The DataFrame containing the model results.
      n (int): The number of top results to display.
      algos (list): A list of tuples, where each tuple contains an algorithm name and an instance of the algorithm.

  Returns:
      None
  """
  for algo_name,algo in algos:
    print('Algorithm Used: ',algo_name)
    print(df[df['Model']==algo_name][['Feature','MSE']].head(n))
    print('='*50)
  del algo_name,algo
  return None
```

```python
# Algorithms to run for our Regression Task
custom_algorithms=[
    ('Linear Regression',LinearRegression()),
    ('Decision Tree',DecisionTreeRegressor(random_state=123)),
    ('Naive Bayes',GaussianNB())
]
```

```python
# Running the model
model_summary=model_run(df=airbnb_df,target_columns=['price'],algos=custom_algorithms,n=0)
print(model_summary.shape)
```

```python
# Displaying top 10 results
display_top_results(df=model_summary,n=10,algos=custom_algorithms)
```
